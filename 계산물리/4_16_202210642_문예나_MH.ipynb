```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = 0.5
x0 = 5
N = 100000

def q(x,m) :
  return np.exp((m-x)/s)*np.exp(-np.exp((m-x)/s))/s

def p(x) :
  return np.exp(-x**2/2)/np.sqrt(2*np.pi)

def al(x,m) :
  return min([1,p(m)*q(x,m)/(p(x)*q(m,x))])
```

```python
# s = 0.5
x = x0
x_list = []
for i in range(N) :
  u = np.random.uniform(0,1)
  m = np.random.gumbel(x,s)
  if u < al(x,m) :
    x = m
  x_list.append(x)

plt.hist(x_list,bins=100)
plt.show()
```
