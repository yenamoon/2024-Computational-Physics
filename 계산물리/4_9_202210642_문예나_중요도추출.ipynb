```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# 5.3.1-(0)
def g1(x) :
  return np.exp((-0.5)*(x**2))

def q1(x) :
  return np.exp(-x)/(1-np.exp(-10))

N = 10000
```

```python
# 5.3.1-(1)
xi = np.random.uniform(0,10,N)
g_arr = np.array(g1(xi))
sum(g_arr)/N
```

```python
# 5.3.1-(2)
u = np.random.uniform(0,1,N)
x = np.array(-np.log(1-(1-np.exp(-10))*u))
#plt.hist(x, bins = 30)
#plt.show()
gw_xi = np.array(g1(x)/q1(x)*0.1)
sum(gw_xi)/N
```

```python
# 5.3.2-(0)
z = 3
N = 10000

def g2(x) :
  return 1 if x > z else 0

def p2(x) :
  return np.exp(-x**2/2)/np.sqrt(2*np.pi)

def q2(x) :
  return np.exp(-(x-z)**2/2)/np.sqrt(2*np.pi)

def w2(x) :
  return p2(x)/q2(x)
```

```python
# 5.3.2-(1)
xi = np.random.normal(0,1,N)
g2_list = list(map(lambda x : g2(x), xi))
sum(g2_list)/N
```

```python
# 5.3.2-(2)
xi = np.random.normal(3,1,N)
gw2_list = list(map(lambda x : g2(x)*w2(x), xi))
sum(gw2_list)/N
```
