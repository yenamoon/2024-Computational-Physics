```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series
```

```python
s = 100
x0 = 5
N = 10000

def q(x,m) :
  return np.exp((m-x)/s)*np.exp(-np.exp((m-x)/s))/s

def p(x) :
  return np.exp(-x**2/2)/np.sqrt(2*np.pi)

def al(x,m) :
  return min([1,p(m)*q(x,m)/(p(x)*q(m,x))])
```

```python
x0 = 5
N = 500
s = 100
x = x0
x_list = []
accept = 0
for i in range(N) :
  u = np.random.uniform(0,1)
  m = np.random.gumbel(x,s)
  if u < al(x,m) :
    x = m
    accept+=1
  x_list.append(x)

print(accept/N)
plt.plot(range(N),x_list)
plt.show()
```
