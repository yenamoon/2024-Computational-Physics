```python
import numpy as np
import matplotlib.pyplot as plt

#p(y)=e^(-y)
N = 10000
x = np.random.uniform(0,1,N)
py = -np.log(x)
plt.hist(py,bins=50)
plt.show()

#정규분포
n = 10
y_arr = x.reshape(n,int(N/n))
y_arr = y_arr.sum(axis=0)
plt.hist(y_arr,bins=50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

dy = 10000
n = 10
N = dy*n

#정규분포/n
n = 10
for _ in range(10) :
  N = dy*n
  x = np.random.uniform(0,1,N)
  y_arr = x.reshape(n,dy)
  y_arr = y_arr.sum(axis=0)/n
  plt.hist(y_arr,bins=50)
  n+=10
plt.show()

#정규분포 n 100배 비교
n = 10
for _ in range(2) :
  N = dy*n
  x = np.random.uniform(0,1,N)
  y_arr = x.reshape(n,dy)
  y_arr = y_arr.sum(axis=0)/n
  plt.hist(y_arr,bins=50)
  n+=990
plt.show()
```
