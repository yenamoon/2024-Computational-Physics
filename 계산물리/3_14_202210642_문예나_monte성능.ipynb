```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
import math

N = 10000
def g(x) :
  return 3*x**2

#(1)uniform
u_total = 0

for _ in range(N) :
  tmp = rd.uniform(0,1)
  u_total += g(tmp)/1

#(2)beta
b_total = 0
a = 2
b = 2

def gam(x) :
  return math.gamma(a+b)/(math.gamma(a)*math.gamma(b))*x**(a-1)*(1-x)**(b-1)

for _ in range(N) :
  tmp = np.random.beta(a,b)
  b_total += g(tmp)/gam(tmp)

#(3)f(y)
f_total = 0

def f(y) :
  return 2*y

for _ in range(N) :
  tmp = rd.uniform(0,1)**0.5
  f_total += g(tmp)/f(tmp)

print('uniform')
print(u_total/N)
print()
print('beta')
print(b_total/N)
print()
print('f(y)')
print(f_total/N)
```

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
import math

N = 10000
def g(x) :
  return 3*x**2

#(1)uniform
u_total = 0
u_10 = []

for _ in range(10) :
  for _ in range(N) :
    tmp = rd.uniform(0,1)
    u_total += g(tmp)/1
  u_10.append(u_total/N)

u_avg = sum(u_10)/10
uq = 0

for i in u_10 :
  uq += (i-u_avg)**2
uq /= N-1


#(2)beta
b_total = 0
b_10 = []
a = 2
b = 2

def gam(x) :
  return math.gamma(a+b)/(math.gamma(a)*math.gamma(b))*x**(a-1)*(1-x)**(b-1)

for _ in range(10) :
  for _ in range(N) :
    tmp = np.random.beta(a,b)
    b_total += g(tmp)/gam(tmp)
  b_10.append(b_total/N)

b_avg = sum(b_10)/10
bq = 0

for i in b_10 :
  bq += (i-b_avg)**2
bq /= N-1

#(3)f(y)
f_total = 0
f_10 = []

def f(y) :
  return 2*y

for _ in range(10) :
  for _ in range(N) :
    tmp = rd.uniform(0,1)**0.5
    f_total += g(tmp)/f(tmp)
  f_10.append(f_total/N)

f_avg = sum(f_10)/10
fq = 0

for i in f_10 :
  fq += (i-f_avg)**2
fq /= N-1

print('uniform')
print(uq)
print()
print('beta')
print(bq)
print()
print('f(y)')
print(fq)
```
