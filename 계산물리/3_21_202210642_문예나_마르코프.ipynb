```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
state = [1,0,0]
S = [1,0,0]
C = [0,1,0]
R = [0,0,1]

def s_func(x) :
  if x < 0.7 :
    return [1,0,0]
  elif x<0.9 :
    return [0,1,0]
  else :
    return [0,0,1]

def c_func(x) :
  if x < 0.3 :
    return [1,0,0]
  elif x<0.95 :
    return [0,1,0]
  else :
    return [0,0,1]

def r_func(x) :
  if x < 0.15 :
    return [1,0,0]
  elif x<0.5 :
    return [0,1,0]
  else :
    return [0,0,1]

value = []
for tmp in range(100) :
  rd_num = np.random.uniform(0,1)

  if state == S :
    value.append(1)
    state = s_func(rd_num)
  elif state == C :
    value.append(2)
    state = c_func(rd_num)
  else :
    value.append(3)
    state = r_func(rd_num)

plt.plot(range(0,100),value)
plt.show()
```
