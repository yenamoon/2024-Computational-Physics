```python
import numpy as np
from functools import reduce
```

```python
# 1
p = 0.4
N = 50
n = 100000

rd = np.random.choice([-1, 1], size=(n, N), p=[1-p, p])
sums = np.sum(rd, axis=1)
count = np.sum(sums > 15)
P = count/n

print(P)
```

```python
# 2
p = 0.4
N = 50
n = 100000
q = 0.65

rd = np.random.choice([-1, 1], size=(n, N), p=[1-q, q])
f = rd[np.sum(rd, axis=1) > 15]
P = np.prod((p/q)**(f == 1) * ((1-p)/(1-q))**(f == -1), axis=1)

result = np.sum(P) / n
result
```
