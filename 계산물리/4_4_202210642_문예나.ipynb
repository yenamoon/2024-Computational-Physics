```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def p_f(x) :
  return 0.3/(1.5*(2*np.pi)**0.5)*np.exp(-0.5*((x+3)/1.5)**2) + 0.7/(1.5*(2*np.pi)**0.5)*np.exp(-0.5*((x-5)/1.5)**2)

n = 10000
```

```python
start = time.time()
x = []
for i in range(n) :
  xi = np.random.uniform(-8,12)
  U = np.random.uniform(0,1)
  if U <= p_f(xi)*5 :
    x.append(xi)
end = time.time()

plt.hist(x, bins=80)
plt.show()
print(end-start)
```

```python
start = time.time()
xi = np.random.uniform(-8,12,n)
U = np.random.uniform(0,1,n)
v = list(map(lambda x,y : x if y <= p_f(x)*5 else 0, xi,U))
v = list(filter(lambda x : x!= 0, v))
end = time.time()

plt.hist(v, bins=80)
plt.show()
print(end-start)
```
