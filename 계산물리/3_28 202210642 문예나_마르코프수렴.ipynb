```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
X = 3
dt = 10000

def s_func() :
  return np.random.choice([1,2,3],p=[0.7,0.2,0.1])

def c_func() :
  return np.random.choice([1,2,3],p=[0.3,0.65,0.05])

def r_func() :
  return np.random.choice([1,2,3],p=[0.15,0.35,0.5])

v_list = []
for _ in range(dt) :
  value = []
  X = 3
  for tmp in range(10) :
    if X == 1 :
      value.append(X)
      X = s_func()
    elif X == 2 :
      value.append(X)
      X = c_func()
    else :
      value.append(X)
      X = r_func()
  v_list.append(value)

v_arr = np.array(v_list)
state_list = []
for i in range(10) :
  state = np.array([0,0,0])
  for j in v_arr[:,i] :
    if j == 1 :
      state[0] += 1
    elif j == 2 :
      state[1] += 1
    else :
      state[2] += 1

  state_list.append(state/dt)

s_arr = np.array(state_list)
plt.plot(range(10),s_arr[:,range(0,3)])
state_list
```

```python
pi = np.array([0,0,1])
P = np.array([[0.7, 0.2, 0.1],[0.3,0.65,0.05],[0.15,0.35,0.5]])

pi_list = []
for i in range(10) :
  pi_list.append(pi)
  pi = pi@P
pi_list = np.array(pi_list)

plt.plot(range(0,10),s_arr)
plt.scatter(range(0,10),pi_list[:,0])
plt.scatter(range(0,10),pi_list[:,1])
plt.scatter(range(0,10),pi_list[:,2])
plt.show()

pi_list
```
